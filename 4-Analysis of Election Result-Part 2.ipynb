```python
# Import relevant libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# Define the information for GE14 and GE15 parties
parties_info = [
    {
        "name": "ge14",
        "alliances": ["PAS", "PH", "BN"],
        "minor_parties": ["Independent", "WARISAN", "STAR"]
    },
    {
        "name": "ge15",
        "alliances": ["PN", "PH", "BN", "GPS", "GRS"],
        "minor_parties": ["Independent", "WARISAN", "MUDA", "PBM", "KDM"]
    }
]

# Iterate over the parties information
for parties in parties_info:
    folder_path = "parties_results"

    # Create a new folder if it doesn't exist
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    # Read the parties result CSV file
    parties_result = pd.read_csv(f"{folder_path}/{parties['name']}_partiesresults.csv")

    # Rename minor parties to "Other"
    parties_result["Alliance"] = parties_result["Alliance"].replace(parties['minor_parties'], "Other")

    # Save the modified parties result to a new CSV file
    parties_result.to_csv(f"{folder_path}/{parties['name']}_partiesresults_modified.csv", index=False)
```

```python
# Define the information for GE14 and GE15 parties
parties_info = [
    {
        "name": "ge14",
        "file_path": "parties_results/ge14_partiesresults_modified.csv"
    },
    {
        "name": "ge15",
        "file_path": "parties_results/ge15_partiesresults_modified.csv"
    }
]

# Initialize an empty dictionary to store party lists for each election
party_lists = {}

# Iterate over the parties information
for parties in parties_info:
    # Read the parties result CSV file
    parties_result = pd.read_csv(parties['file_path'])

    # Get the unique party names for the current election
    unique_parties = parties_result['Alliance'].unique().tolist()

    # Store the unique party names in the dictionary
    party_lists[parties['name']] = unique_parties

# Print the party lists for each election
for election, parties in party_lists.items():
    print(f"Party List for {election.upper()}:")
    for party in parties:
        print(party)
    print()
```

```python
# Define the information for GE14 and GE15 parties
parties_info = [
    {
        "name": "ge14",
        "file_path": "parties_results/ge14_partiesresults_modified.csv",
        "custom_colors": {
            "PAS": "#009000",
            "PH": "#ED1C24",
            "BN": "#000080",
            "Other": "#A4E5FC"
        }
    },
    {
        "name": "ge15",
        "file_path": "parties_results/ge15_partiesresults_modified.csv",
        "custom_colors": {
            "PN": "#031e61",
            "PH": "#ED1C24",
            "BN": "#000080",
            "PAS": "#009000",
            "GRS": "#4682B4",
            "GPS": "#FF6060",
            "Other": "#A4E5FC"
        }
    }
]

# Loop through the parties information
for parties in parties_info:
    # Read the parties result CSV file
    parties_result = pd.read_csv(parties['file_path'])

    # Initialize an empty dictionary to store the seat counts for each party
    seat_counts = {}

    # Loop through the parties
    for party in parties_result['Alliance'].unique():
        seat_counts[party] = len(parties_result[parties_result['Alliance'] == party])

    # Create a new figure
    plt.figure(figsize=(6, 12))

    # Define custom colors for parties
    custom_colors = [parties['custom_colors'].get(party, '#E8E8E8') for party in seat_counts.keys()]

    # Plotting data on chart
    plt.pie(seat_counts.values(), labels=seat_counts.keys(), colors=custom_colors,
            autopct=lambda p: f'{p:.0f}% ({p * sum(seat_counts.values()) / 100:.0f})')

    # Add title and legend
    plt.title(f"Alliance-based Seat Distribution in Malaysia's {parties['name'].upper()} General Election")
    plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
   
    # Conditionally add minor parties information in text annotation
    if parties['name'] == 'ge14':
        plt.text(0, -1.2, '(Other: Independent, WARISAN, STAR)', ha='right', fontsize=7)
    elif parties['name'] == 'ge15':
        plt.text(0, -1.2, '(Other : KDM, PBM, WARISAN, MUDA, Independent)', ha='right', fontsize=7)
    
    # Save the figure as a PNG image
    plt.savefig(f"election_analysis/{parties['name']}_seat_distribution.png", dpi=300, bbox_inches='tight')

    # Display the chart
    plt.show()
```

```python
# Define the information for GE14 and GE15 parties
parties_info = [
    {
        "name": "ge14",
        "file_path": "parties_results/ge14_partiesresults_modified.csv",
        "alliance": ["PAS", "PH", "BN"],
        "custom_colors": {
            "PAS": "#009000",
            "PH": "#ED1C24",
            "BN": "#000080"
        }
    },
    {
        "name": "ge15",
        "file_path": "parties_results/ge15_partiesresults_modified.csv",
        "alliance": ["PN", "PH", "BN"],
        "custom_colors": {
            "PN": "#031e61",
            "PH": "#ED1C24",
            "BN": "#000080"
        }
    }
]
```

```python
# Iterate over the parties info
for parties in parties_info:
    # Read the parties result CSV file
    parties_result = pd.read_csv(parties['file_path'])
    # Get the current alliances for the loop iteration
    alliances = parties['alliance']
    # Get the current custom colors
    custom_colors = parties['custom_colors']
    
    # Create a new figure
    fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(6, 12))
    
    # Iterate over the alliances
    for i, alliance in enumerate(alliances):
        num_bins = 20
        # Filter the data for the current alliance
        alliance_data = parties_result[parties_result['Alliance'] == alliance]
        # Get the margin-of-votes values for the histogram
        margin_of_votes = alliance_data["Margin-of-Votes"]
        # Plot the histogram on the corresponding subplot
        axes[i].hist(margin_of_votes, bins=num_bins, color=custom_colors[alliance])
        # Set the y-axis limit for the subplot
        axes[i].set_ylim([0, 15])
        # Set the x-axis limit for the subplot
        axes[i].set_xlim([0, 90])
        # Set the title for the subplot
        axes[i].set_title(alliance)
        axes[i].axvline(x=6.0, color='k', linestyle='--')
        axes[i].axvline(x=10.0, color='k', linestyle='--')
        axes[i].axvline(x=19.0, color='k', linestyle='--')
        # Add labels for the boundary lines
        axes[i].text(3.2, 12.7, 'Marginal', ha='right', va='center', rotation=90, fontsize=8, color='k')
        axes[i].text(9.5, 12.7, 'Fairly Safe', ha='right', va='center', rotation=90, fontsize=8, color='k')
        axes[i].text(18.5, 12.7, 'Safe', ha='right', va='center', rotation=90, fontsize=8, color='k')
        axes[i].text(25, 12.7, 'Very Safe', ha='right', va='center', rotation=90, fontsize=8, color='k')
    
    # Add a common x-axis label
    fig.text(0.5, -0.01, 'Margin-of-Votes (%)', ha='center', fontsize=12)
    # Add a common y-axis label
    fig.text(0.01, 0.5, 'Seats', va='center', rotation='vertical', fontsize=12)
    
    # Add a suptitle for the entire figure
    fig.suptitle(f'Seat Distribution by Margin-of-Votes - {parties["name"].upper()}', fontsize=14)
    
    # Adjust the spacing between subplots
    plt.tight_layout()
    
    # Save the figure as a PNG image
    plt.savefig(f'election_analysis/Seat Distribution by alliances histogram - {parties["name"].upper()}.png', dpi=300, bbox_inches='tight')
    
    # Show the plot
    plt.show()
```

```python
color_classifications = [
    {"color": "#7D7D7D"},
    {"color": "#D3D3D3"},
    {"color": "#9E9E9E"},
    {"color": "#696969"}
]
```

```python
# Iterate over the parties info
for parties in parties_info:
    # Read the parties result CSV file
    parties_result = pd.read_csv(parties['file_path'])
    # Get the current alliances for the loop iteration
    alliances = parties['alliance']
    # Get the current custom colors
    custom_colors = [color['color'] for color in color_classifications]
    
    # Create subplots with one row and three columns
    fig, axes = plt.subplots(nrows=len(alliances), ncols=1, figsize=(6, 12))

    # Iterate through the alliances
    for i, alliance in enumerate(alliances):
        # Filter the data for the current alliance
        alliance_data = parties_result[parties_result["Alliance"] == alliance]

        # Calculate the seat counts for each margin-of-votes classification
        counts = alliance_data["Classification"].value_counts()

        # Format the autopct parameter to include percentages and number of seats
        def autopct_format(pct):
            value = pct / 100 * counts.sum()
            return f"{pct:.0f}% ({value:.0f})"

        # Plot the pie chart on the corresponding subplot
        patches, texts, autotexts = axes[i].pie(counts, colors=custom_colors, autopct=autopct_format)
        axes[i].set_title(alliance, fontsize=13)
    
    # Add a legend to the figure
    fig.legend(patches, counts.index, loc="center", bbox_to_anchor=(0.5, -0.02), ncol=len(counts))
    
    # Add a common title for the entire figure
    fig.suptitle(f"Seat Distribution by Margin-of-Votes Classification - {parties['name'].upper()}", fontsize=14)

    # Adjust the spacing between subplots
    plt.tight_layout()

    # Save the figure as a PNG image
    plt.savefig(f'election_analysis/Seat Distribution by alliances piechart - {parties["name"].upper()}.png', dpi=300, bbox_inches='tight')

    # Show the plot
    plt.show()
```

```python
for parties in parties_info:
    # Read the parties result CSV file
    parties_result = pd.read_csv(parties['file_path'])
    # Get the current alliances for the loop iteration
    alliances = parties['alliance']
    
    sns.set_theme(style="whitegrid", palette="muted")

    # Filter the data for the three alliances
    alliance_data = parties_result[parties_result["Alliance"].isin(alliances)]

    # Create a swarmplot for all three alliances
    ax = sns.swarmplot(data=alliance_data, x="Margin-of-Votes", y="Alliance", hue="Classification")
    ax.text(6, 0.5, 'Marginal', ha='right', va='center', rotation=90, fontsize=8, color='k')
    ax.text(10, 0.5, 'Fairly Safe', ha='right', va='center', rotation=90, fontsize=8, color='k')
    ax.text(19, 0.5, 'Safe', ha='right', va='center', rotation=90, fontsize=8, color='k')
    ax.text(24, 0.5, 'Very Safe', ha='right', va='center', rotation=90, fontsize=8, color='k')
    ax.axvline(x=6.0, color='k', linestyle='--')
    ax.axvline(x=10.0, color='k', linestyle='--')
    ax.axvline(x=19.0, color='k', linestyle='--')

    # Set the y-axis label
    ax.set(ylabel="", xlabel="Margin (%)")

    # Add a title to the plot
    plt.title("Seat Distribution by Margin-of-Votes Classification - " + parties['name'].upper())

    # Save the figure as a PNG image
    plt.savefig(f"election_analysis/{parties['name'].lower()}_seat_distribution_swarmplot.png", dpi=300, bbox_inches='tight')

    # Show the plot
    plt.show()
```

```python
import pandas as pd
```

```python
ge15_result = pd.read_csv('parties_results/ge15_partiesresults.csv')
ge14_result = pd.read_csv('parties_results/ge14_partiesresults.csv')
```

```python
ge14_result.head()
```

```python
# Specify the columns of interest
columns_of_interest = ['Parliament Code', 'Federal Constituency', 'Alliance', 'Margin-of-Votes']

# Merge the two dataframes based on 'Parliament Code' and 'Federal Constituency' columns
merged_df = pd.merge(ge14_result[columns_of_interest], ge15_result[columns_of_interest], on=['Parliament Code', 'Federal Constituency'])

# Specify custom column names for the new dataframe
new_column_names = ['Parliament Code', 'Federal Constituency', 'GE14_Alliance', 'GE14_Margin', 'GE15_Alliance', 'GE15_Margin']

# Assign the custom column names to the merged dataframe
merged_df.columns = new_column_names

merged_df.head()
```

```python
# Sort the merged dataframe based on 'Parliament Code' in ascending order
sorted_df = merged_df.sort_values('Parliament Code')

# Filter rows where Alliance is not changing and GE15_Margin is lower than GE14_Margin
lower_margin_condition = (sorted_df['GE14_Alliance'] == sorted_df['GE15_Alliance']) & (sorted_df['GE15_Margin'] < sorted_df['GE14_Margin'])
Seat_Hold_LowerMargin = sorted_df[lower_margin_condition]

# Filter rows where Alliance is not changing and GE15_Margin is higher than GE14_Margin
higher_margin_condition = (sorted_df['GE14_Alliance'] == sorted_df['GE15_Alliance']) & (sorted_df['GE15_Margin'] > sorted_df['GE14_Margin'])
Seat_Hold_HigherMargin = sorted_df[higher_margin_condition]

# Save Seat_Hold_LowerMargin dataframe as CSV
Seat_Hold_LowerMargin.to_csv('election_analysis/Seat_Hold_LowerMargin.csv', index=False)

# Save Seat_Hold_HigherMargin dataframe as CSV
Seat_Hold_HigherMargin.to_csv('election_analysis/Seat_Hold_HigherMargin.csv', index=False)
```
